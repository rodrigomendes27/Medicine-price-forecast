## Label Encording

```python
from sklearn.preprocessing import LabelEncoder

le = dict()
cat_cols = ["CAP","descrição"]
for col in cat_cols:
  le[col] = LabelEncoder()
  df_2_copy[col] = le[col].fit_transform(df_2_copy[col])
```

## Divisão dos dados de treino e teste

```python
from sklearn.model_selection import train_test_split
label = df_2_copy["correto"]
feat_cols = df_2_copy.columns.tolist()
feat_cols.remove("correto")
feat = df_2_copy[feat_cols]

X_train, X_test, y_train, y_test = train_test_split(feat, label, random_state=1)
```

```python
df_2_copy.info()
```

```python
X_train.data[50]
```

```python
# Transformando data de String para Integer

from datetime import datetime

X_train["data"] = X_train["data"].apply(lambda d: datetime.strptime(d, "%Y-%m-%d"))
X_train["Year"] = X_train["data"].apply(lambda d: d.year)
X_train["Month"] = X_train["data"].apply(lambda d: d.month)
X_train["Day"] = X_train["data"].apply(lambda d: d.day)
X_train["time_as_customer"] = X_train["data"].apply(lambda d: (datetime.now() - d).days)
X_train.drop(columns=["data"], inplace=True)


X_test["data"] = X_test["data"].apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
X_test["Year"] = X_test["data"].apply(lambda d: d.year)
X_test["Month"] = X_test["data"].apply(lambda d: d.month)
X_test["Day"] = X_test["data"].apply(lambda d: d.day)
X_test["time_as_customer"] = X_test["data"].apply(lambda d: (datetime.now() - d).days)

X_test.drop(columns=["data"], inplace=True)
```

# Normalização dos dados

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_ = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

# Modelagem

## Regressão Linear

```python
from sklearn import metrics
import numpy as np
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_, y_train)
test_pred_lin = lin_reg.predict(X_test)
```

```python
# Validação

smape_lin = 100/len(y_test) * np.sum(2 * np.abs(test_pred_lin - y_test) / (np.abs(y_test) + np.abs(test_pred_lin)))
rmse_lin = metrics.mean_squared_error(y_test, test_pred_lin, squared=False)
r2_lin = metrics.r2_score(y_test, test_pred_lin)


print(f"SMAPE da Regressão Linear: {smape_lin}")
print(f"RMSE da Regressão linear: {rmse_lin}")
print(f"R² da Regressao linear: {r2_lin}")
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor()
rf_reg.fit(X_, y_train)
test_pred_rf = rf_reg.predict(X_test)
```

```python
# Validação

smape_rf = 100/len(y_test) * np.sum(2 * np.abs(test_pred_rf - y_test) / (np.abs(y_test) + np.abs(test_pred_rf)))
rmsle_rf = np.sqrt(metrics.mean_squared_log_error(y_test, test_pred_rf))
rmse_rf = metrics.mean_squared_error(y_test, test_pred_rf, squared=False)
r2_rf = metrics.r2_score(y_test, test_pred_rf)


print(f"SMAPE do Random Forest: {smape_rf}")
print(f"RMSLE do Random Forest: {rmsle_rf}")
print(f"RMSE do Random Forest: {rmse_rf}")
print(f"R² do Random Forest: {r2_rf}")
```

## XGBoost

```python
import xgboost as xgb

clf_XGB = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=15,
    scale_pos_weight=100_000,
    use_label_encoder=True
)
clf_XGB.fit(X_, y_train)
```

```python
test_pred_xg = clf_XGB.predict(X_test)
```

```python
# Validação

smape_rf = 100/len(y_test) * np.sum(2 * np.abs(test_pred_xg - y_test) / (np.abs(y_test) + np.abs(test_pred_xg)))
rmsle_rf = np.sqrt(metrics.mean_squared_log_error(y_test, test_pred_xg))
rmse_rf = metrics.mean_squared_error(y_test, test_pred_xg, squared=False)
r2_rf = metrics.r2_score(y_test, test_pred_xg)


print(f"SMAPE do Random Forest: {smape_rf}")
print(f"RMSLE do Random Forest: {rmsle_rf}")
print(f"RMSE do Random Forest: {rmse_rf}")
print(f"R² do Random Forest: {r2_rf}")
```
